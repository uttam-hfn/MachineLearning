```python
# Import package
import seaborn as sns
```

```python
# Load data
tips = sns.load_dataset('tips')
tips.columns = ['Total bill', 'Tip', 'Sex', 'Smoker', 'Day', 'Time', 'Party size']
tips
```

```python
# Create scatter plot with color
sns.scatterplot(data=tips, x='Total bill', y='Tip', hue='Smoker', style='Smoker')
```

```python
# Change opacity with alpha
sns.scatterplot(
    data=tips, x='Total bill', y='Tip', hue='Smoker', style='Smoker', alpha=0.7
)
```

```python
# Change color palette with palette
sns.scatterplot(
    data=tips,
    x='Total bill',
    y='Tip',
    hue='Smoker',
    style='Smoker',
    alpha=0.7,
    palette='colorblind',
)
```

```python
# Faceted scatter plot
sns.relplot(
    data=tips,
    x="Total bill",
    y="Tip",
    col="Sex",
    hue="Sex",
    size="Party size",
    style="Time",
    height=3,
)
```

```python
# Faceted count plot
sns.catplot(
    data=tips,
    x="Time",
    hue="Smoker",
    col="Day",
    col_wrap=2,
    kind="count",
    height=2.5,
    aspect=0.6,
)
```
