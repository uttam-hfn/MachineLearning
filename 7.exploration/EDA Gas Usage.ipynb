## Data description

Date : The day for the energy usage

Gas : The amount of natural gas used in therms

Low : The low temperature for the day in Fahrenheit

Average : The average temperature for the day in Fahrenheit

High : The high temperature for the day in Fahrenheit

SolarkWh : The amount of electricity generated by the solar panels in kWh

FireHrs : Duration of the fire in the woodstove in hours

AwayHrs : Number of hours the thermostat was setback to the away temperature

PanelsFree : The proportion of the solar panels that is not covered with snow

ConsumptionkWh : The amount of electricity consumed by the household in kWh

HotWaterkWh : The amount of electricity consumed by the heat pump water heater in kWh

Guests: Whether guests stayed at the house. (0 = no guests, 1 = guests visited)
