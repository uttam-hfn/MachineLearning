```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Configuration
D = 2 # Vector Dimesion, we usually use 2 x,y
K = 3 # number of clusterings
N = 300 # number of data points
```

```python
# means
means = np.array([
    [0, 0],
    [0, 5],
    [5, 5]
])
```

```python
X = np.random.randn(N, D)
X[:100,:] = np.random.randn(100, D) + means[0]
X[100:200,:] = np.random.randn(100, D) + means[1]
X[200:,:] = np.random.randn(100, D) + means[2]
```

```python
Y  = np.zeros(N)
for n in range(N):
    closest_k = -1
    min_dist = float('inf')
    for k in range(K):
        d = (X[n] - means[k]).dot(X[n] - means[k])
        if d < min_dist:
            min_dist = d
            closest_k = k
        Y[n] = closest_k
```

```python
# visualize the data
plt.scatter(X[:,0], X[:,1], c=Y)
```
