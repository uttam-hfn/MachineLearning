```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create features
X = np.array([-7.0, -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0])

# Create labels
y = np.array([3.0, 6.0, 9.0, 12.0, 15.0, 18.0, 21.0, 24.0])

# Visualize it
plt.scatter(X, y);
```

```python
# Set random seed
tf.random.set_seed(42)
```

```python
# Create a model using the Sequential API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
# Compile the model
model.compile(loss=tf.keras.losses.mae, # mae is short for mean absolute error
              optimizer=tf.keras.optimizers.SGD(), # SGD is short for stochastic gradient descent
              metrics=["mae"])

# Fit the model
model.fit(tf.expand_dims(X, axis=-1), y, epochs=1000)
```

```python
# Check out X and y
X, y
```

```python
# Make a prediction with the model
prediction = model.predict(np.array([[17.0]]))
prediction
```

```python
# Make a bigger dataset
X = np.arange(-100, 100, 4)
X
```

```python
y = X**2
y
```

```python
plt.scatter(X, y);
```

```python
# Define the model using the Sequential API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation=tf.keras.activations.relu),
    tf.keras.layers.Dense(4, activation=tf.keras.activations.relu),
    tf.keras.layers.Dense(1, activation=None)  # No activation on the output layer for regression
])

# Compile the model
model.compile(
    loss=tf.keras.losses.mse,  # Use mean squared error for regression tasks
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.08),
    metrics=["mae"])  # Use mean absolute error as an additional metric

# Fit the model
history = model.fit(tf.expand_dims(X, axis=-1), y, epochs=1000)
```

```python
# Split data into train and test sets
X_train = X[:35] # first 40 examples (80% of data)
y_train = y[:35]

X_test = X[35:] # last 10 examples (20% of data)
y_test = y[35:]

len(X_train), len(X_test)
```

```python
plt.figure(figsize=(10, 7))
# Plot training data in blue
plt.scatter(X_train, y_train, c='b', label='Training data')
# Plot test data in green
plt.scatter(X_test, y_test, c='g', label='Testing data')
# Show the legend
plt.legend();
```

```python
# Make predictions
y_preds = model.predict(X_test)
```

```python
y_preds
```

```python
def plot_predictions(train_data=X_train,
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                     predictions=y_preds):
    """
    Plots training data, test data and compares predictions.
    """
    plt.figure(figsize=(10, 7))
    # Plot training data in blue
    plt.scatter(train_data, train_labels, c="b", label="Training data")
    # Plot test data in green
    plt.scatter(test_data, test_labels, c="g", label="Testing data")
    # Plot the predictions in red (predictions were made on the test data)
    plt.scatter(test_data, predictions, c="r", label="Predictions")
    # Show the legend
    plt.legend();
```

```python
plot_predictions(train_data=X_train,
                 train_labels=y_train,
                 test_data=X_test,
                 test_labels=y_test,
                 predictions=y_preds)
```

```python
model.evaluate(X_test, y_test)
```
